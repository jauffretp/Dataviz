On regarde l'age en fonction des amendements adoptés

```python
def arrondi(row):
    if row.age < 30:
        return 25
    elif row.age < 40:
        return 35
    elif row.age < 50:
        return 45
    elif row.age < 60:
        return 55
    elif row.age < 70:
        return 65
    else:
        return 75
    
activite_depute['age_arrondi'] = activite_depute.apply(arrondi, axis=1)
```

```python
activite_depute[['age_arrondi','amendements_adoptes','hemicycle_interventions_courtes']].groupby(by='age_arrondi').mean()
```

# Twitter

```python
import pandas as pd
depute_twitter = pd.read_csv("twitter\\deputes.csv", sep=",", index_col=1)
```

```python
def arrondi(row):
    if row.twitter_tweets < 10000:
        return 5000
    elif row.twitter_tweets < 20000:
        return 15000
    elif row.twitter_tweets < 30000:
        return 25000
    elif row.twitter_tweets >= 30000:
        return 35000
    
depute_twitter['nb_tweets_arrondi'] = depute_twitter.apply(arrondi, axis=1)
```

```python
activite_depute = pd.read_csv("nosdeputes.fr_synthese_2017-02-22.csv", sep=",", index_col=2)

depute_twitter = depute_twitter.merge(activite_depute, how='inner', left_index=True, right_index=True)
```

```python
import pandas as pd

import json

import io
```

```python
tweets = pd.read_csv("tweets.csv", sep=",", encoding="UTF-8")
depute = pd.read_csv("nosdeputes.fr_synthese_2017-02-22.csv", sep=",", encoding="UTF-8")
collab = pd.read_csv("liste_deputes_collaborateurs.csv", sep=",", encoding="UTF-8")
merged = pd.merge(tweets, depute, on="slug", how="inner")
groupby = merged[["twitter_followers", "twitter_tweets", "parti_ratt_financier", "nom_y"]].groupby("parti_ratt_financier")
```

```python
"éol".encode("UTF-8")
```

```python

datasets = []
names = []
for name,group in groupby:
    dobject = {}
    dobject["label"] = name
    data = []
    for index,row in group.iterrows():
        robject = {}
        robject["r"] = 5
        robject["x"] = row["twitter_followers"]
        robject["y"] = row["twitter_tweets"]
        data.append(robject)
    
    dobject["data"] = data
    datasets.append(dobject)
    names.append(group["nom_y"].tolist())
```

```python
with io.open("final_output.js", "w", encoding="UTF-8") as file:
    file.write(json.dumps(datasets,indent=4,ensure_ascii=False))
    file.write("\n\n\n")
    file.write(json.dumps(names, indent=4,ensure_ascii=False))
```
