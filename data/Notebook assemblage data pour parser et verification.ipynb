```python
import pandas as pd

import json

import io
```

```python
tweets = pd.read_csv("tweets.csv", sep=",", encoding="UTF-8")
depute = pd.read_csv("nosdeputes.fr_synthese_2017-02-22.csv", sep=",", encoding="UTF-8")
collab = pd.read_csv("liste_deputes_collaborateurs.csv", sep=",", encoding="UTF-8")
merged = pd.merge(tweets, depute, on="slug", how="inner")
groupby = merged[["twitter_followers", "twitter_tweets", "parti_ratt_financier", "nom_y"]].groupby("parti_ratt_financier")
```

```python
"éol".encode("UTF-8")
```

```python

datasets = []
names = []
for name,group in groupby:
    dobject = {}
    dobject["label"] = name
    data = []
    for index,row in group.iterrows():
        robject = {}
        robject["r"] = 5
        robject["x"] = row["twitter_followers"]
        robject["y"] = row["twitter_tweets"]
        data.append(robject)
    
    dobject["data"] = data
    datasets.append(dobject)
    names.append(group["nom_y"].tolist())
```

```python
with io.open("final_output.js", "w", encoding="UTF-8") as file:
    file.write(json.dumps(datasets,indent=4,ensure_ascii=False))
    file.write("\n\n\n")
    file.write(json.dumps(names, indent=4,ensure_ascii=False))
```
